```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines 

%matplotlib inline
```

## 这里主要是定义函数以及起始点变量

```python
def f(x):
    return (x ** 3) + (4 * x ** 2) - 10
```

```python
a = float(-1)
b = float(2)
limit = float(0.65)

x = np.linspace(-1, 2, 1000)
y = f(x)
```

## 下面是绘图块

```python
plt.figure()
plt.xlabel('x')
plt.ylabel('y')
# 设定左右轴
ax = plt.gca()
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.plot(x, y)
plt.show()
```

```python
def draw(a: float, b: float):
    ax.add_line(mlines.Line2D([a, f(a)], [b, f(b)], color='red'))
    plt.draw()
```

```python
from time import sleep
def calcu_pos(a: float, b: float, limit: float):
    xr = b - ((f(b) * (a - b)) / (f(a) - f(b)))
    while f(xr) != 0 and abs(a - b) >= limit:
        # print(str(xr))
        sleep(1)
        if f(a) * f(xr) > 0:
            a = xr
        else:
            b = xr
        draw(a, b)
        xr = b - ((f(b) * (a - b)) / (f(a) - f(b)))
        
        # print(str(abs(a - b)))
    return xr
calcu_pos(a, b, limit)
plt.show()
```
